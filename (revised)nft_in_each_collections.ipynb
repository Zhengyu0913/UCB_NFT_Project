```python
import pandas as pd

collections_df = pd.read_csv('collection_appear_4.csv', index_col=[0])
collections_df
```

```python
s_name=[] 

for s in range(len(collections_df['symbol'])):
    s_name.append(collections_df['symbol'][s])

s_name
```

```python
col_list = ['id','price', 'owner', 'collectionName','collectionTitle','title','propertyCategory','creators','attributes','properties', 'tokenDelegateValid','rarity'
]

v1_col_list = col_list = ['id','price', 'owner', 'collectionName','collectionTitle','title','propertyCategory','creators','attributes','properties', 'tokenDelegateValid'
]
```

```python
v1_missing_property_list=[]
v2_missing_property_list=[]
v3_missing_property_list=[]
v4_missing_property_list=[]
```

```python
# create a for loop to get all collection symbol under symbol column
# sep='' : with no space front/back of input value
li_v1=[]

for s in range(len(s_name)):
    li_v1.append("solana_nft_full/solana_v1/Solana_NFT_Collections_v1/collections/"+s_name[s]+".csv")

v1_list = []

for x in range(len(li_v1)):
    try:
        df = pd.read_csv(li_v1[x], usecols=v1_col_list)
        v1_list.append(df)
    except Exception as e:
        print(e)
        print("cannot: ", li_v1[x])
        v1_missing_property_list.append(li_v1[x])
        pass
        

v1_df = pd.concat(v1_list, axis=0, ignore_index=True)
v1_df['rarity'] = "NaN"
v1_df['version'] = "1"
v1_df
```

```python
li_v2=[]

for s in range(len(s_name)):
    li_v2.append("solana_nft_full/solana_v2/Solana_NFT_Collections_v2/collections/"+s_name[s]+".csv")

v2_list = []

for x in range(len(li_v2)):
    try:
        df = pd.read_csv(li_v2[x], usecols=col_list)
        v2_list.append(df)
    except Exception as e:
        print(e)
        print("cannot: ", li_v2[x])
        v2_missing_property_list.append(li_v2[x])
        pass
        

v2_df = pd.concat(v2_list, axis=0, ignore_index=True)
v2_df['version'] = "2"
v2_df
```

```python
li_v3=[]

for s in range(len(s_name)):
    li_v3.append("solana_nft_full/solana_v3/Solana_NFT_Collections_v3/collections/"+s_name[s]+".csv")

v3_list = []

for x in range(len(li_v3)):
    try:
        df = pd.read_csv(li_v3[x], usecols=col_list)
        v3_list.append(df)
    except Exception as e:
        print(e)
        print("cannot: ", li_v3[x])
        v3_missing_property_list.append(li_v3[x])
        pass
        

v3_df = pd.concat(v3_list, axis=0, ignore_index=True)
v3_df['version'] = "3"
v3_df
```

```python
li_v4=[]

for s in range(len(s_name)):
    li_v4.append("solana_nft_full/solana_v4/Solana_NFT_Collections_v4/collections/"+s_name[s]+".csv")

v4_list = []

for x in range(len(li_v4)):
    try:
        df = pd.read_csv(li_v4[x], usecols=col_list)
        v4_list.append(df)
    except Exception as e:
        print(e)
        print("cannot: ", li_v4[x])
        v4_missing_property_list.append(li_v4[x])
        pass
        

v4_df = pd.concat(v4_list, axis=0, ignore_index=True)
v4_df['version'] = "4"
v4_df
```

```python
nopro_col_list = ['id','price', 'owner', 'collectionName','collectionTitle','title','creators','attributes','properties', 'tokenDelegateValid','rarity'
]

v1_nopro_col_list = ['id','price', 'owner', 'collectionName','collectionTitle','title','creators','attributes','properties', 'tokenDelegateValid']

v1_nonpro_list=[]
v2_nonpro_list=[]
v3_nonpro_list=[]
v4_nonpro_list=[]
```

```python
for x in range(len(v1_missing_property_list)):
    try:
        df = pd.read_csv(v1_missing_property_list[x], usecols=v1_nopro_col_list)
        v1_nonpro_list.append(df)
    except Exception as e:
        print(e)
        print("cannot: ", v1_missing_property_list[x])
        pass
    
v1_nonpro_list
```

```python
for x in range(len(v2_missing_property_list)):
    try:
        df = pd.read_csv(v2_missing_property_list[x], usecols=nopro_col_list)
        v2_nonpro_list.append(df)
    except Exception as e:
        print(e)
        print("cannot: ", v2_missing_property_list[x])
        pass
    
v2_nonpro_list
```

```python
for x in range(len(v3_missing_property_list)):
    try:
        df = pd.read_csv(v3_missing_property_list[x], usecols=nopro_col_list)
        v3_nonpro_list.append(df)
    except Exception as e:
        print(e)
        print("cannot: ", v3_missing_property_list[x])
        pass
    
v3_nonpro_list
```

```python
for x in range(len(v4_missing_property_list)):
    try:
        df = pd.read_csv(v4_missing_property_list[x], usecols=nopro_col_list)
        v4_nonpro_list.append(df)
    except Exception as e:
        print(e)
        print("cannot: ", v4_missing_property_list[x])
        pass
    
v4_nonpro_list
```

```python
nopro_v1_df = pd.concat(v1_nonpro_list, axis=0, ignore_index=True)
nopro_v1_df.insert(6,'propertyCategory',"NaN")
nopro_v1_df['rarity'] = "NaN"
nopro_v1_df['version'] = "1"

nopro_v2_df = pd.concat(v2_nonpro_list, axis=0, ignore_index=True)
nopro_v2_df.insert(6,'propertyCategory',"NaN")
nopro_v2_df['version'] = "2"

nopro_v3_df = pd.concat(v3_nonpro_list, axis=0, ignore_index=True)
nopro_v3_df.insert(6,'propertyCategory',"NaN")
nopro_v3_df['version'] = "3"

nopro_v4_df = pd.concat(v4_nonpro_list, axis=0, ignore_index=True)
nopro_v4_df.insert(6,'propertyCategory',"NaN")
nopro_v4_df['version'] = "4"
```
